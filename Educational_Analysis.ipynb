```python
import pandas as pd

df = pd.read_csv('higher_education_india.csv')
print(df.head())

# Example: Plot GER by State
import matplotlib.pyplot as plt

df.plot(x='State', y='GER', kind='bar', figsize=(12,6), title='Gross Enrollment Ratio by State')
plt.ylabel('GER (%)')
plt.xlabel('State')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
